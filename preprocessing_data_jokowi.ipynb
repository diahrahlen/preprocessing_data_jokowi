# Preprocessing Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
import nltk
import re
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
```

```python
#masukan file yg akan di preprocessing#

df = pd.read_csv('file_jokowi.csv', encoding = 'utf-8')
df
```

1) Tahap Normalisasi atau Cleaning

```python
#menghapus user pada kolom text

def remove_pattern(text, pattern):
    r = re.findall(pattern, str(text))
    for i in r:
        text = re.sub(i, '', str(text))
    return text
df['remove_user'] = np.vectorize(remove_pattern)(df['text'], "@[\w]*")
df
```

```python
#menghapus karakter yang tidak penting pada kolom text

def remove_punct(text):
    text = re.sub(r'\$\w*', '', text)
    text = re.sub(r'^rt[\s]+', '', text)
    text = re.sub('((www\.[^\s]+)|(https?://[^\s]+))', ' ', text)
    text = re.sub('&quot;'," ", text)
    text = re.sub(r"\d+", " ", str(text))
    text = re.sub(r"\b[a-zA-Z]\b", "", str(text))
    text = re.sub(r"[^\w\s]", " ", str(text))
    text = re.sub(r'(.)\1+', r'\1\1', text)
    text = re.sub(r"\s+", " ", str(text))
    text = re.sub(r'#', '', text)
    text = re.sub(r'[^a-zA-z0-9]', ' ', str(text))
    text = re.sub(r'\b\w{1,2}\b', '', text)
    text = re.sub(r'\s\s+', ' ', text)
    text = re.sub(r'^RT[\s]+', '', text)
    text = re.sub(r'^b[\s]+', '', text)
    text = re.sub(r'^link[\s]+', '', text)
   
    return text

df['cleaning'] = df['remove_user'].apply(lambda x: remove_punct(x))
df
```

2) Tahap Case Folding

```python
#mengubah huruf besar menjadi huruf kecil (terutama pada kolom text, remove_user, cleaning)

df['text'] = df['text'].str.lower()
df['remove_user'] = df['remove_user'].str.lower()
df['cleaning'] = df['cleaning'].str.lower()
df
```

3) Tahap Tokenization

```python
#membagi text menjadi potongan-potongan kecil

def tokenization(text):
    text = re.split('\W+', text)
    return text

df['tokenisasi'] = df['cleaning'].apply(lambda x: tokenization(x.lower()))
df
```

4) Tahap Stopword Removal

```python
#menghapus kata yg tidak ada artinya

stopword = nltk.corpus.stopwords.words('Indonesian')

def remove_stopwords(text):
    text = [word for word in text if word not in stopword]
    return text

df['stopword_removal'] = df['tokenisasi'].apply(lambda x: remove_stopwords(x))
df
```

```python
#menghapus karakter [] dan koma

stop_removal = df[['stopword_removal']]

def fit_stopwords(text):
    text = np.array(text)
    text = ' '.join(text)
    
    return text

df['stopword_removal'] = df['stopword_removal'].apply(lambda x: fit_stopwords(x))
df
```

5) Filtering text duplicate

```python
#menghapus text duplicate dengan menetapkan text pertama

df.drop_duplicates(subset = "stopword_removal", keep = 'first', inplace = True)
df
```

```python
#simpan file dalam bentuk csv

df.to_csv('databersih_file_jokowi.csv',encoding='utf8', index=False)
```
